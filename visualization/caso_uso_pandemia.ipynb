## La gran pandemia

Durante el 2015 hubo una gran pandemia a nivel mundial. Luego de un año del descubrimiento del virus, se realizó un catastro de enfermos. Cómo es posible observar en el siguiente gráfico, los resultados indicaron que los países más afectados fueron Francia, Japón y México, con más de 9.5 millones, 10.2 millones y 10.3 millones de contagiados respectivamente.


**Visualización 1 (V01)**

**Tarea identificada**: Acción/Objetivo = Descubrir - Extremos;

Acción = Analizar(Descubrir), Objetivos = Atributos (Extremos).

**Justificación**: En esta visualización corresponde utilizar un gráfico de barras tradicional puesto que se requiere graficar en el eje X el país (que vendría siendo la variable categórica) y en el eje Y la variable cuantitativa que correspondería Completar en máximo 5 líneas.

```python
# V01: Visualización que permite ver la cantidad total de infectados por país
# Dataset: V01.csv 

#librerias necesarias
import matplotlib.pyplot as plt
from matplotlib import font_manager as fm
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
# Utilizar la función read_csv para cargar el archivo
v01 = pd.read_csv("V01.csv")


v01['Contagiados_mm'] = v01['Contagiados'].div(1000000).round(1)
v01
```

```python
# graficamos ahora el arreglo resultante
ax = v01.plot.bar(color = 'green', title = 'Contagiados por país', x = 'Paises', y = 'Contagiados_mm')
ax.get_legend().remove()
ax.set_xlabel("País")
ax.set_ylabel("Contagiados (MM)")
```

Explorando los datos históricos de esta pandemia, fue posible recopilar la información de cantidad de contagiados internacionalmente durante los primeros 5 años y los resultados mostraron un crecimiento exponencial de los casos.

**Visualización 2 (V02)**

**Tarea identificada**: Acción/Objetivo = Presentar - Tendencias

Acción: Analizar (Consume -> Present), Objetivo: All Data -> Tendencias.

**Justificación**: Puesto que la idea en esta visualización es mostrar los datos evolutivos durante los primeros 5 años y mostrar que hubo un crecimiento en los casos es que se requiere del uso del gráfico de puntos y de líneas puesto que se utiliza principalmente para presentar tendencias, en donde el eje X sería la variable ordinal (período) y el eje Y la cantidad (contagios), utilizaremos el de líneas para poder dar la sensación de continuidad.

```python
# V02: Visualización que permita ver la evolución del total de infectados en el tiempo (fecha).
# Dataset: V02.csv

# Utilizar la función read_csv para cargar el archivo
v02 = pd.read_csv("V02.csv")

# Ver las primeras 5 líneas del archivo
v02.head(5)
```

Ahora, comparando la situación por país, es posible notar cómo en algunos países, como Chile, la infección fue controlada mientras en otros no, como México. Esto se explica por la anticipada toma de decisiones en salud realizados en varios de estos países.


**Visualización 3 (V03)**

**Tarea identificada**: Acción/Objetivo = Comparar - Tendencias

Acciones: Consultar -> Comparar, Objetivo: All Data -> Tendencia.

**Justificación**: En este caso se requiere ver la evolución temporal de la pandemia pero por cada país, en este caso necesitaremos el uso del canal de color para poder identificar por la variable categórica del país, sin embargo si utilizamos todo en el mismo gráfico quedaría poco intuitivo y claro para el lector, por lo que la mejor opción de visualización es un gráfico de Pequeños múltiples ya que todos los gráficos comparten los mismos atributos y cada gráfico posee solo 1 categoría posible.

```python
# V03: Visualización que permita ver el total de infectados en el tiempo por país (fecha).
# Dataset: V03.csv 

# Utilizar la función read_csv para cargar el archivo
v03 = pd.read_csv("V03.csv")

# Ver las primeras 5 líneas del archivo
v03.head(5)
```

```python
v03_melt = pd.melt(v03, id_vars =['Pais'], value_vars =['2013', '2014','2015','2016','2017']) 
v03_melt = v03_melt.rename(columns={'variable': 'Año', 'value':'Contagiados'})

v03_melt.head(5)
```

```python
# Generar una grilla utilizando como variable el género
grid = sns.FacetGrid(data=v03_melt, col="Pais", hue="Pais", col_wrap=4)
# Llamamos al método map que se encarga de llenar cada gráfico de la grilla utiliznado año en el eje X y cantidad en el Y
grid.map(plt.plot, "Año", "Contagiados", marker="D")

# Visualizar el gráfico
plt.show()
```

Las visualizaciones anteriores colocan la evolución del contagios según las fechas de ocurrencia, pero la realidad es que cada país tuvo su primer infectado en distintos momentos. Al encontrarse en distintas escalas de tiempo, surge la interrogante de cómo distribuyen los crecimientos de contagios bajo una misma escala de tiempo, que corresponde al tiempo que lleva el país con infecciones presentes. **Visualización 6 (V06)**

**Tarea identificada**: Acción/Objetivo = Descubrir - Distribución

**Justificación**: En este caso es necesario un tipo de visualización que permita ver las proporciones de los contagiados según variables categóricas que son la edad y el tipo de síntoma, para este caso lo que más nos sirve para poder representarlo es el gráfico de tortao pie chart, puesto que permite visualizar proporciones en base a variables categóricas.

```python
# V06: Dos visualizaciones que permitan mostrar la proporción de infectados por edad y la proporción de síntomas posibles.
# Dataset: V06.csv 

# Utilizar la función read_csv para cargar el archivo
v06 = pd.read_csv("V06.csv")
# Ver las primeras 5 líneas del archivo
v06
```

```python
# graficamos ahora el arreglo resultante
v06_group_edad = v06.groupby("Edad").count()
v06_group_sintomas = v06.groupby("Sintomas").count()

v06_group_edad
```

```python
#Creacion de subplots para ubicar los gráficos de torta
fig, (ax1, ax2) = plt.subplots(1, 2)
fig.set_dpi(100)

# Create two subplots and unpack the output array immediately
ax1.pie(v06_group_edad["Sintomas"], autopct='%1.1f%%',labels =  v06_group_edad.index)
ax2.pie(v06_group_sintomas["Edad"], autopct='%1.1f%%',labels =  v06_group_sintomas.index)

ax1.set(aspect="equal", title='Proporción de contagiados según Edad')
ax2.set(aspect="equal", title='Proporción de contagiados según Síntomas')

#Se asigna un título al gráfico
fig.suptitle("Proporción de contagios según variables de Edad y Síntomas")

# Visualizar el gráfico
plt.show()
```
